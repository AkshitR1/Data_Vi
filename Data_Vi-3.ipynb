```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
df_can = pd.read_excel("C:\\Users\\rtran\\Downloads\\Canada.xlsx",
    sheet_name='Canada by Citizenship',
    skiprows=range(20),
    skipfooter=2)
all(isinstance(column, str) for column in df_can.columns)
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)
years = list(map(str, range(1980, 2014)))
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage', 'Region', 'DevName'], axis=1, inplace=True)
df_can.head()
```

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
df_continents.head()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

df_continents[2001].plot(kind='pie', figsize=(13,13),
                            autopct='%1.f%%', # add in percentages
                            startangle=90,     # start angle 90° (Africa)
                            )
plt.title='pie'
plt.show()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
df_09 = df_can[2009].transpose()
df_09.plot(kind='box', figsize=(8, 6))
plt.title('Box plot of Immigrants from 2009')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
# we can use the sum() method to get the total population per year
df_tot = pd.DataFrame(df_continents.sum(axis=0))

# change the years to type int (useful for regression later on)
df_tot.index = map(int, df_tot.index)

# reset the index to put in back in as a column in the df_tot dataframe
df_tot.reset_index(inplace = True)

# rename columns
df_tot.columns = ['year', 'Continents']

# view the final dataframe
df_tot.tail()
```

```python
df_tot.plot(kind='scatter', x='year', y='Continents', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants from various continents')

plt.show()
```
