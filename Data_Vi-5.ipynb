```python
india_map = folium.Map(location=[20 , 77], zoom_start=5)
#adding marker
Delhi = folium.map.FeatureGroup()
Delhi.add_child(folium.features.CircleMarker([28.38,	77.17], radius=5, color='red', fill_color='Red'))
india_map.add_child(Delhi)
folium.Marker([28.38,	77.17], popup='Delhi').add_to(india_map)
#Mumbai
Mumbai = folium.map.FeatureGroup()
Mumbai.add_child(folium.features.CircleMarker([18.55, 72.50], radius=5, color='blue', fill_color='Blue'))
india_map.add_child(Mumbai)
folium.Marker([18.55,	72.50], popup='Mumbai').add_to(india_map)

#Bengaluru
Bengaluru = folium.map.FeatureGroup()
Bengaluru.add_child(folium.features.CircleMarker([12.59, 77.40], radius=5, color='orange', fill_color='Orange'))
india_map.add_child(Bengaluru)
folium.Marker([12.59, 77.40], popup='Bengaluru').add_to(india_map)

#Hyderabad
Hyderabad = folium.map.FeatureGroup()
Hyderabad.add_child(folium.features.CircleMarker([17.22,78.29], radius=5, color='black', fill_color='White'))
india_map.add_child(Hyderabad)
folium.Marker([17.22,78.29], popup='Hyderabad').add_to(india_map)

#Kolkata
Kolkata = folium.map.FeatureGroup()
Kolkata.add_child(folium.features.CircleMarker([22.36, 88.24], radius=5, color='brown', fill_color='Yellow'))
india_map.add_child(Kolkata)
folium.Marker([22.36, 88.24], popup='Kolkata').add_to(india_map)

india_map
```
